```python
# ==========================================
# TUGAS AI: KNOWLEDGE BASE EKOSISTEM
# Kelompok 1
# ==========================================

class SistemPakarLingkungan:
    def __init__(self):
        # 1. KNOWLEDGE BASE (Basis Pengetahuan)
        # Sesuai persis dengan Semantic Network dari gambar draw.io terbaru
        self.kb = {
            # --- Bagian Ekosistem ---
            "Matahari": [("Mempengaruhi", "Iklim"), ("Menyinari/Memberikan energi", "Pohon")],
            "Tanah": [("Mendukung", "Pohon")],
            "Air": [("Menyuburkan", "Pohon")],
            "Iklim": [("Mempengaruhi", "Hutan")],
            "Hutan": [
                ("Bagian Dari", "Ekosistem"),
                ("Menyerap", "Co2"),
                ("Adalah habitat bagi", "Hewan")
            ],
            "Hewan": [
                ("Menghasilkan", "Co2"),
                ("Membutuhkan", "Oksigen")
            ],
            "Manusia": [
                ("Menghasilkan", "Co2"),
                ("Membutuhkan", "Oksigen")
            ],
            "Co2": [],       # Entitas pasif di diagram ini
            "Oksigen": [],   # Entitas pasif di diagram ini
            "Ekosistem": [], # Entitas ujung di diagram ini

            # --- Bagian Hubungan Ekosistem & Anatomi ---
            "Pohon": [
                ("Bagian Dari", "Hutan"),
                ("Menghasilkan", "Oksigen"),
                ("Memiliki", "Batang"),
                ("Memiliki", "Akar")
            ],

            # --- Bagian Anatomi Pohon ---
            "Batang": [("Memiliki", "Ranting")],
            "Ranting": [
                ("Memiliki", "Buah"),
                ("Memiliki", "Daun"),
                ("Memiliki", "Bunga")
            ],
            "Akar": [],
            "Buah": [],
            "Daun": [],
            "Bunga": []
        }

    # 2. MESIN INFERENSI (Fungsi untuk mencari dan menyimpulkan)
    def cari_fakta(self, entitas):
        # Normalisasi teks agar pencarian akurat (misal: "pohon" -> "Pohon")
        entitas = entitas.capitalize()

        print(f"\n[  MENGANALISIS ENTITAS: {entitas.upper()} ]")

        if entitas in self.kb:
            relasi_list = self.kb[entitas]

            if not relasi_list:
                print(f"  ➔ '{entitas}' adalah entitas akhir/pasif dalam diagram ini.")
                return

            print("Fakta yang ditemukan:")
            for relasi, target in relasi_list:
                print(f"  ➔ {entitas} {relasi} {target}")

            # Logika Cerdas Tambahan (Membaca Anatomi Pohon secara rekursif)
            if entitas == "Pohon":
                print("\n  [ Info Tambahan Anatomi]:")
                print("    Karena Pohon memiliki Batang, dan Batang memiliki Ranting, maka:")
                print("    ➔ Pohon secara tidak langsung menghasilkan Buah, Daun, dan Bunga.")

            elif entitas == "Hutan":
                print("\n  [ Info Tambahan Ekosistem]:")
                print("    Hutan sangat penting karena ia menyerap Co2 yang dihasilkan oleh Manusia dan Hewan.")

        else:
            print(f" Entitas '{entitas}' tidak ada di dalam database.")

# ==========================================
# 3. DEMO PROGRAM (Jalankan di bawah ini)
# ==========================================

# Inisialisasi AI
ai = SistemPakarLingkungan()

print("=== SISTEM AI LINGKUNGAN KELOMPOK 1 ===")
print("Memuat Knowledge Base (16 Entitas)...\n")

# Menjalankan test case sesuai kelompok
ai.cari_fakta("Matahari")
ai.cari_fakta("Hutan")
ai.cari_fakta("Pohon")
ai.cari_fakta("Ranting")
ai.cari_fakta("Manusia")
```
