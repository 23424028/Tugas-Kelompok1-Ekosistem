```python
# ==========================================
# KNOWLEDGE BASE
# ==========================================

knowledge_base = [

    ("Hutan", "bagian_dari", "Ekosistem"),
    ("Pohon", "bagian_dari", "Hutan"),

    ("Matahari", "mempengaruhi", "Iklim"),
    ("Iklim", "mempengaruhi", "Pohon"),

    ("Tanah", "mendukung", "Pohon"),
    ("Air", "menyuburkan", "Pohon"),

    ("Pohon", "menyerap", "CO2"),
    ("Pohon", "menghasilkan", "Oksigen"),

    ("Hutan", "habitat", "Hewan"),
    ("Hewan", "membutuhkan", "Oksigen"),
    ("Hewan", "menghasilkan", "CO2"),

    ("Manusia", "membutuhkan", "Oksigen"),
    ("Manusia", "menghasilkan", "CO2"),

    ("CO2", "digunakan_oleh", "Pohon"),
    ("Oksigen", "digunakan_oleh", "Hewan"),
    ("Oksigen", "digunakan_oleh", "Manusia"),

    ("Pohon", "memiliki", "Batang"),
    ("Pohon", "memiliki", "Akar"),
]

# ==========================================
# DESKRIPSI
# ==========================================

deskripsi = {
    "pohon": "Tumbuhan yang berperan sebagai produsen dan menghasilkan oksigen melalui fotosintesis.",
    "hewan": "Makhluk hidup konsumen yang membutuhkan oksigen.",
    "manusia": "Makhluk hidup yang bergantung pada lingkungan dan membutuhkan oksigen.",
    "oksigen": "Gas yang dibutuhkan makhluk hidup untuk bernapas.",
    "co2": "Gas karbon dioksida yang digunakan tumbuhan untuk fotosintesis.",
    "hutan": "Ekosistem yang menjadi habitat berbagai makhluk hidup."
}

# ==========================================
# FUNGSI INTI
# ==========================================

def cari_relasi(entitas):
    hasil = []
    for s, p, o in knowledge_base:
        if s.lower() == entitas:
            hasil.append(f"{s} {p} {o}")
        elif o.lower() == entitas:
            hasil.append(f"{s} {p} {o}")
    return hasil

# ==========================================
# NLP SEDERHANA (RULE BASED)
# ==========================================

while True:
    pertanyaan = input("\nTanya: ").lower()

    if pertanyaan == "exit":
        break

    # ==========================
    # APA ITU (DESKRIPSI)
    # ==========================
    if "apa itu" in pertanyaan:
        entitas = pertanyaan.replace("apa itu", "").strip()

        if entitas in deskripsi:
            print(deskripsi[entitas])
        else:
            print("Maaf, penjelasan tidak ditemukan.")

    # ==========================
    # RELASI
    # ==========================
    elif "relasi" in pertanyaan:
        kata = pertanyaan.split()
        entitas = kata[0]  # ambil kata pertama

        hasil = cari_relasi(entitas)

        if hasil:
            print(f"Relasi {entitas}:")
            for h in hasil:
                print("-", h)
        else:
            print("Tidak ada relasi ditemukan.")

    # ==========================
    # DEFAULT
    # ==========================
    else:
        print("Pertanyaan tidak dikenali. Contoh:")
        print("- apa itu pohon")
        print("- pohon relasi dengan apa")
```
